```python
def fib(n):    # write Fibonacci series up to n
    a, b = 0, 1
    text_file = open("Output.txt", "w")
    while b < n:
        print(b)
        text_file.write(str(b))
        text_file.write("\n")
        a, b = b, a+b
    text_file.close()   
print("write Fibonacci series up to n")
n = int(input("n="))
fib(n)
```
